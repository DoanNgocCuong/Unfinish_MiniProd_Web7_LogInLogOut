        <div className="App">

          <Routes>
            <Route 
              path="/" 
              element={
                <Home 
                  message={message} 
                  t={t} 
                  users={users} 
                  departments={departments} 
                >
                  {/* Add ToolsGrid inside Home */}
                  <ToolsGrid />

              } 
            />
            {/* Other routes */}
          </Routes>

          {location.pathname === '/' && <Footer />}

    </ChakraProvider>
  );
}
